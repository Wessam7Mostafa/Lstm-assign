```python
"""LSTM assignment .ipynb

Automatically generated by Colab.

Original file is located at
    https://colab.research.google.com/drive/1SRET9Ha30gs44SbYZLuy5DiZiIlSLp4W
"""

import numpy as np

def sigmoid(x):
    return 1/(1 + np.exp(-x))

def tanh(x):
    return np.tanh(x)

sequence = [1, 2, 3]

Wf, Whf, bf = 0.5, 0.1, 0
Wi, Whi, bi = 0.6, 0.2, 0
Wc, Whc, bc = 0.7, 0.3, 0
Wo, Who, bo = 0.8, 0.4, 0

h_t = 0
C_t = 0

print("Step by Step LSTM Calculation\n")

for step in range(len(sequence)):

    x_t = sequence[step]

    forget = sigmoid(Wf * x_t + Whf * h_t + bf)

    inp = sigmoid(Wi * x_t + Whi * h_t + bi)

    candidate = tanh(Wc * x_t + Whc * h_t + bc)

    C_t = forget * C_t + inp * candidate

    out = sigmoid(Wo * x_t + Who * h_t + bo)

    h_t = out * tanh(C_t)

    print("Time:", step+1)
    print("Input:", x_t)
    print("f =", round(forget,3))
    print("i =", round(inp,3))
    print("C~ =", round(candidate,3))
    print("C =", round(C_t,3))
    print("o =", round(out,3))
    print("h =", round(h_t,3))
    print("-----------------------")

next_input = 4

f = sigmoid(Wf * next_input + Whf * h_t + bf)
i = sigmoid(Wi * next_input + Whi * h_t + bi)
C_bar = tanh(Wc * next_input + Whc * h_t + bc)

C_t = f * C_t + i * C_bar

o = sigmoid(Wo * next_input + Who * h_t + bo)
h_t = o * tanh(C_t)

print("\nPrediction Step")
print("Input:", next_input)
print("f =", round(f,3))
print("i =", round(i,3))
print("C~ =", round(C_bar,3))
print("C =", round(C_t,3))
print("o =", round(o,3))
print("h =", round(h_t,3))

Wy = 4
by = 0

y_hat = Wy * h_t + by
print("Predicted value =", round(y_hat,3))

```
